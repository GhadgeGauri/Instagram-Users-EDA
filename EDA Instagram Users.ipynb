```python
def min_max_val(col):
    top=instagram[col].idxmax()
    top_obs=pd.DataFrame(instagram.loc[top])
    
    bottom=instagram[col].idxmin()
    bottom_obs=pd.DataFrame(instagram.loc[bottom])
    
    min_max_obs=pd.concat([top_obs, bottom_obs],axis=1)
    return min_max_obs
```

```python
min_max_val('Followers')
```
